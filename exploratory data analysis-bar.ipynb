```python
import pandas as pd
import matplotlib.pyplot as plt

# Exploratory Data Analysis (EDA) - Bar Chart 10 Most Brand
df = pd.read_csv("Automobile.csv")

# Extract 'merk' (brand) from the first word of the 'name' column
df['merk'] = df['name'].str.split().str[0]

# Count Occurences
top_context = df['merk'].value_counts().head(10) 

# Bar Chart
plt.figure(figsize=(10, 6))
ax = top_context.plot(kind='bar')
# Add label
for i, value in enumerate(top_context):
    plt.text(i, value + 1, str(value), ha='center', va='bottom')
plt.title("10 Most Brand")
plt.xlabel("Brand")
plt.ylabel("Number of Car")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Exploratory Data Analysis (EDA) - Top 10 Brands by Highest Average Horsepower
df = pd.read_csv("Automobile.csv")

# Extract 'merk' (brand) from the first word of the 'name' column
df['merk'] = df['name'].str.split().str[0]

# Convert 'horsepower' to numeric and clean empty value
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
df = df.dropna(subset=['horsepower'])

# Group by brand
max_hp_per_brand = df.loc[df.groupby('merk')['horsepower'].idxmax()]

# Get top 10 brands by highest horsepower
top_hp_brands = max_hp_per_brand.sort_values(by='horsepower', ascending=False).head(10)

# Bar Chart
plt.figure(figsize=(10, 6))
ax = plt.bar(top_hp_brands['merk'], top_hp_brands['horsepower'])
# Add label
for i, value in enumerate(top_hp_brands['horsepower']):
    plt.text(i, value + 2, f"{value:.0f}", ha='center', va='bottom')
plt.title("Top 10 Brands by Highest Average Horsepower")
plt.xlabel("Brand")
plt.ylabel("Horsepower")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Exploratory Data Analysis (EDA) - Stacked Bar Chart per Origin by Weight Category
df = pd.read_csv("Automobile.csv")

# Convert 'weight' to numeric and clean empty value
df['weight'] = pd.to_numeric(df['weight'], errors='coerce')
df = df.dropna(subset=['weight'])

# Create weight categories
bins = [1000, 2000, 3000, float('inf')]
labels = ['1000-2000', '2001-3000', '3000>']
df['weight_category'] = pd.cut(df['weight'], bins=bins, labels=labels, right=True, include_lowest=True)

# Group by origin and weight category
grouped = df.groupby(['origin', 'weight_category'], observed=False).size().unstack(fill_value=0)

# Stacked Bar Chart
plt.figure(figsize=(10, 6))
bottom = pd.Series([0] * len(grouped), index=grouped.index)

for category in labels:
    values = grouped[category] if category in grouped.columns else pd.Series([0] * len(grouped), index=grouped.index)
    bars = plt.bar(grouped.index, values, bottom=bottom, label=category)
    # Add value label in center of each segment
    for i, value in enumerate(values):
        if value > 0:
            plt.text(i, bottom.iloc[i] + value / 2, f"{int(value)}", ha='center', va='center', color='white', fontsize=9)
    bottom += values
plt.title("Stacked Weight Category Distribution per Origin")
plt.xlabel("Origin")
plt.ylabel("Count")
plt.legend(title="Weight Category")
plt.tight_layout()
plt.show()
```
