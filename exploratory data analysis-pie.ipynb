```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Automobile.csv")

# Exploratory Data Analysis (EDA) - Pie Chart Most Origin
top_context = df['origin'].value_counts().head(7)

# Pie Chart
plt.figure(figsize=(8,8))
top_context.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title(f"Most Origin")
plt.ylabel("")
plt.show()
```
